Separando os dados

```python
X=clientes.iloc[:,3:7]
Y=clientes['Yearly Amount Spent']
```

Divisão dos Dados em Treino e Teste

```python
from sklearn.model_selection import train_test_split
```

```python
Xtreino,Xteste,Ytreino,Yteste=train_test_split(X,Y,test_size=0.3,random_state=123)
```

```python
Xtreino.shape, Xteste.shape
```

Modelo de Regressão Linear

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(Xtreino, Ytreino)
```

```python
lm.coef_
```

Previsões com os dados de Teste

```python
Yprev=lm.predict(Xteste)
```

```python
Yprev
```

```python
plt.scatter(Yteste, Yprev)
plt.xlabel('Y REAL')
plt.ylabel('Y PREVISTO')
plt.show()
```

```python
Yteste
```

```python
 Yprev
```

```python
#Comparação dos Valores
pd.DataFrame({'Valor Real':Yteste,'Previsão':Yprev})
```

```python
# Cálculo do R2
#98% dos dados é explicado pelo modelo de regressão
from sklearn.metrics import r2_score
R2=r2_score(Yteste,Yprev)
R2
```

Algumas métricas para verificação do modelo.

MAE: Mean Absolute error é a média do erro, diferença do valor que foi previsto pelo seu modelo e o valor real que foi observado.

MSE: Erro Quadrático Médio consiste na média do erro das previsões ao quadrado.

RMSE:Raiz Quadrada do Erro Quadrático Médio.

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(Yteste, Yprev))
print('MSE:', metrics.mean_squared_error(Yteste, Yprev))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Yteste, Yprev)))
```
